properties of normal distribution:
 1.mean,median,mode are all equal and they are at their peak and symmetrical in nature(mean divides into two equal parts)
 2.68% if data lies within the first standard deviation.
 3.95% of data lies within the second standard deviation.
 4.99% of data lies within third standard deviation.
 
 5.first standard deviation lies b/w (mean - sd) to (mean + sd)
 6.second standard deviation lies b/w (mean - 2sd) to (mean + 2sd)
 7.third standard deviation lies b/w (mean - 3sd) to (mean + 3sd)

```python
import random
import plotly.express as px  
import plotly.figure_factory as ff
import statistics


dice_result = []
count=[]
for i in range(0,1000):
    dice1 = random.randint(1,6)
    #print(dice1)
    dice2 = random.randint(1,6)
    #print(dice2)
    dice_sum = dice1+dice2
    #print(dice_sum)
    dice_result.append(dice_sum)
    #print(dice_result)
    count.append(i)

marks = [80,20,30,60,60,40,40,50,60,30]
m = statistics.median(marks)
print(m)
mean = statistics.mean(dice_result)

print("mean of the data is {}".format(mean) )

standard_deviation = statistics.stdev(dice_result)
print("standard deviation of the data is {}".format(standard_deviation) )

median = statistics.median(dice_result)
print("median of the data is {}".format(median) )

mode = statistics.mode(dice_result)
print("mode of the data is {}".format(mode) )

first_sd_start,first_sd_end = mean - standard_deviation,mean + standard_deviation
listofdata_within_1stStDeviation=[result for result in dice_result if result > first_sd_start and result < first_sd_end]
print(listofdata_within_1stStDeviation)

print("{}% of data lies within 1st standard deviation ".format(len(listofdata_within_1stStDeviation)*100/len(dice_result)))

second_sd_start,second_sd_end = mean - 2*standard_deviation,mean + 2*standard_deviation
listofdata_within_2ndStDeviation=[result for result in dice_result if result > second_sd_start and result < second_sd_end]
print(listofdata_within_2ndStDeviation)

print("{}% of data lies within 2nd standard deviation ".format(len(listofdata_within_2ndStDeviation)*100/len(dice_result)))

third_sd_start,third_sd_end = mean - 3*standard_deviation,mean + 3*standard_deviation
listofdata_within_3rdStDeviation=[result for result in dice_result if result > third_sd_start and result < third_sd_end]
print(listofdata_within_3rdStDeviation)

print("{}% of data lies within 3rd standard deviation ".format(len(listofdata_within_3rdStDeviation)*100/len(dice_result)))
```
